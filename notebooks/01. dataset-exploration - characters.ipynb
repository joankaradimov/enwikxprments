```python
with open('page_revisions_text', 'rb') as text_file:
    data = text_file.read()
```

```python
articles = data.split(b'\0')
del data
```

```python
characters = [0] * 256
for article in articles:
    for character in article:
        characters[character] += 1

characters[0] = len(articles)
```

```python
print('Unique characters in the dataset:', sum(1 if c > 0 else 0 for c in characters))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20, 15))
plt.bar(range(256), characters);
```

```python
lowcase_articles = [article.lower() for article in articles]

lowcase_articles_characters = [0] * 256
for article in lowcase_articles:
    for character in article:
        lowcase_articles_characters[character] += 1

lowcase_articles_characters[0] = len(lowcase_articles)
```

```python
plt.figure(figsize=(20, 15))
plt.bar(range(256), lowcase_articles_characters);
```

```python
plt.figure(figsize=(20, 15))
plt.bar(range(128), lowcase_articles_characters[:128]);
```

```python
print('Unique characters in the dataset:', sum(1 if c > 0 else 0 for c in lowcase_articles_characters[:128]))
```

```python
articles_length = [len(a) for a in articles]

plt.figure(figsize=(20, 15))
plt.hist(articles_length, bins=50);
```

```python
print('Max article length:', max(articles_length))
```
