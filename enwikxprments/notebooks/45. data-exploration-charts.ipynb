Интересно е да се види дали дисперсията намалява, докато превиждането напредва...

```python
def pareto_variance(predictions):
    cdf = np.array(sorted(predictions, reverse=True)).cumsum()
    pareto_fit = pareto.fit(cdf)
    return pareto(*pareto_fit).var()
    
variances = [pareto_variance(predicted_chars[i]) for i in range(1000)]
variances
```

```python
plt.figure(figsize=(5, 4))
plt.margins(0.003)
plt.plot(range(1000), [v if v < 10 else 10 for v in variances])
plt.yscale('linear', vmax=10)
plt.show();
```
